```python

## Saving the data to a text   
with open("Output.txt", "w") as text_file:
    text_file.write(data)

## Removing stopwords
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
def remove_stop_words(data):
  
  word_tokens = word_tokenize(data)
  
  filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
  
  filtered_sentence = []
  
  for w in word_tokens:
      if w not in stop_words:
          filtered_sentence.append(w)
  return filtered_sentence

filtered_sentence = remove_stop_words(data)
```

```python

# Counting word frequency
def wordFrequency(list_of_words):
    word_frequencies = {}
    for word in list_of_words:
        word_frequencies[word.lower()] = 1 + word_frequencies.get(word.lower(), 0)
    # Sort the dictionary by value and return it
    return sorted(word_frequencies.items(), key = lambda word: word[1], reverse = True )

word_count_mapped = wordFrequency(filtered_sentence)


word_count_mapped
```

```python
# Problem 2.2
# Let Z be a randomly selected word in a randomly selected ICML paper.  Estimate the entropy of Z.
import numpy as np
import random


# Distribution = [1,2,3] is of different probability
def entropy_log_function(fraction):
    if (fraction == 0.0):
        return 0.0
    ent = -fraction * np.log2(fraction)
    if ent == 0.0:
      return 0.0
    return (ent)

def entropy_func(list_of_probabilities):
  entropy_sum = 0
  for probability in list_of_probabilities:
    entropy_sum += entropy_log_function(probability)
  return entropy_sum


# Replace full_words with actual words
#Calculating  of a randomly selected word probability of words : 
def prob_distribution(clean_data, total_word_count):
  prob_words = []
  for word in clean_data:
    prob_words.append(word[1]/total_word_count)
  return prob_words

print("Entropy")

probability_distribution_z = prob_distribution(clean_data, total_words)
print(entropy_func(prob_distribution(clean_data, total_words)))
```
