Wrtie a program that on input k and XXXX (year), returns the top k names...

```python
def kpopular (k,year):
  filename='yob'+str(year)+'.txt'
  with open('/content/'+filename,'r') as file:
    file = file.readlines()
  lines = file[:k]
  names = []
  for line in lines:
    names.append(line.split(",")[0])
  return names
```

# 3.b
Write a program that on input Name, returns the frequency for men and women of the Name...

```python
all_names = []

for year in range(1880, 2016):
    all_names.append(pd.read_csv('./Names/yob{}.txt'.format(year), names=['name', 'sex', 'number']))
    all_names[-1]['year'] = year

allyears_name = pd.concat(all_names)
allyears_name.head()
```

```python
import pandas as pd
def getFrequency (name):
  df=pd.DataFrame(allyears_name)
  df_name=df.loc[df['name'] ==name]
  female_count=df_name.loc[df_name['sex']=='F'].number.sum()
  male_count=df_name.loc[df_name['sex']=='M'].number.sum()
  print("men: " + str(male_count))
  print("women: " + str(female_count))
  # Returns the frequency of men and women of the name
  return [male_count, female_count]

getFrequency('Mary') 
```

```python
# Get the relative frequency which is frequency of it happening / total outcomes
import pandas as pd
def getRelativeFrequency (name, year):
  df=pd.DataFrame(allyears_name)
  df_rel=pd.read_csv('/content/yob{}.txt'.format(year), names=['name', 'sex', 'number'])
  df_name=df.loc[df['name'] ==name]
  df_name_rel = df_rel.loc[df_rel['name'] ==name]
  female_count=df_name.loc[df_name['sex']=='F'].number.sum()
  female_count_rel=df_name_rel.loc[df_name_rel['sex']=='F'].number.sum()
  male_count=df_name.loc[df_name['sex']=='M'].number.sum()
  male_count_rel=df_name_rel.loc[df_name_rel['sex']=='M'].number.sum()
  rel_freq_male = male_count_rel / male_count
  rel_freq_female = female_count_rel / female_count
  print("men frequency compared from " + str(year) + " to all years: " + str(rel_freq_male))
  print("women frequency compared from " + str(year) + " to all years: " + str(rel_freq_female))
  # Returns the frequency of men and women of the name
  return [rel_freq_male, rel_freq_female]

getRelativeFrequency('Mary', 2010) 
```

```python
change_trends=pd.DataFrame(columns=['name','male_popular','female_popular'])
for year in range(1880, 2016):
    names_y=pd.read_csv('/content/yob{}.txt'.format(year), names=['name', 'sex', 'frequency'])
    names_m= names_y.loc[names_y['sex']=='M',['name','frequency']]
    names_m.rename(columns={'frequency':'male_freq'},inplace=True)

    names_f= names_y.loc[names_y['sex']=='F',['name','frequency']]
    names_f.rename(columns={'frequency':'female_freq'},inplace=True)

    collab_names=names_f.merge(names_m,how='outer',on='name')

    collab_names.fillna(0,inplace=True)
    collab_names['male_popular']=collab_names['male_freq']>collab_names['female_freq']
    collab_names['female_popular']=collab_names['female_freq']>collab_names['male_freq']

    name_trends=collab_names[['name','male_popular','female_popular']].set_index(["name"])
    change_trends.update(name_trends,filter_func=lambda x:x==False)

    change_trends=pd.concat([change_trends,name_trends]).reset_index().drop_duplicates(['name'],keep='first').set_index(['name'])
    # print(change_trends)
    change_trends_n=change_trends[change_trends['male_popular']& change_trends['female_popular']].sort_values(by='name')
    change_trends_n.reset_index(inplace=True)
print(change_trends_n['name'])





```
