1. Create 1000 samples from a Gaussian distribution with mean -10 and standard deviation 5. Create another 1000 samples from another independent Gaussian with mean 10 and standard deviation 5.

(a)  Take the sum of these Gaussians by adding the two sets of 1000 points,  point by point, and plot the histogram of the resulting 1000 points.  What do you observe?

Deliverables: three histograms (two for each Gaussian, one for the sum), written response, code

(b)  Estimate the mean and the variance of the sum.
Deliverables: written response, code

```python
import numpy as np
from matplotlib import pyplot as plt

# Part 1 - Gaussian is normal distribution, so we need to use numpy.random.normal(mean, standard deviation, size)
p1_1000_samples1 = np.random.normal(-10, 5, 1000) 
p1_1000_samples2 = np.random.normal(10, 5, 1000) 
```

```python
# Part 1a - Take the sum of these gaussians by adding the two sets of 1000 points, point by point and plot the histogram of the resulting 1000 points
#p1a_1000_samples_sum = p1_1000_samples1 + p1_1000_samples2
p1a_1000_samples_sum = []
for x in range(0, 1000):
  p1a_1000_samples_sum.append(p1_1000_samples1[x] + p1_1000_samples2[x])

# First plot
mu1, sigma1 = -10, 5
count, bins, ignored=plt.hist(p1_1000_samples1, bins=30, density=True, alpha=0.6, color='b')
plt.plot(bins, 1/(sigma1 * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu1)**2 / (2 * sigma1**2) ),
         linewidth=2, color='r')
plt.title("1st 1000 Gaussian Distribution")
plt.show()

# Second plot
mu2, sigma2 = 10, 5
count, bins, ignored=plt.hist(p1_1000_samples2, bins=30, density=True, alpha=0.6, color='b')
plt.plot(bins, 1/(sigma2 * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu2)**2 / (2 * sigma2**2) ),
         linewidth=2, color='r')
plt.title("2nd 1000 Gaussian Distribution")
plt.show()

# Sum plot
plt.hist(p1a_1000_samples_sum)
plt.title("Sum of 2 histograms")
plt.show()
```

# 1A
The first graph clearly shows that the majority of points huddle towards -10 (x-axis), which is its mean. The majority of non -10 points, spread out about +/- 5 (x-axis) before hitting less than 1/2 the height of the chart (as is the standard deviation).

The second graph is kind of similar to the first, just transposed +20 on the x-axis. It clearly shows that the majority of points huddle towards 10 (x-axis), which is its mean. The majority of non 10 points, spread out about +/- 5 (x-axis) before hitting less than 1/2 the height of the chart (as is the standard deviation).

The final (third) graph appears to be not only transposed so the majority of points lie around x=0, but it is also a wider graph (as you can see by the separation of points on the x axis). The standard deviation here looks more like 10 than 5.

Generate 25,000 samples from a Gaussian distribution with mean 0 and standard deviation 5.

```python
p2_25000_samples1 = np.random.normal(0, 5, 25000) 
```

```python
# Then estimate the mean and standard deviation of this gaussian using elementary numpy commands,  i.e.,  addition, multiplication,  division 
#(do  not  use  a  command  that  takes  data  and  returns  the  mean  or standard deviation).

# mean - add all points together and divide by total count
p2_mean = np.sum(p2_25000_samples1)/np.size(p2_25000_samples1)
print("mean: " + str(p2_mean))

# standard deviation
# find the distance from each point to the average
p2_distance_list = []
for point in p2_25000_samples1:
  p2_distance_list.append((p2_mean - point) ** 2)
# convert to array
p2_distance_arr = np.array(p2_distance_list)
# get the sum of the array
p2_distance_sum = np.sum(p2_distance_arr)
# divide by total number of points
p2_distance_divided = p2_distance_sum / np.size(p2_distance_arr)
# get the standard deviation by getting its square root
p2_standard_deviation = p2_distance_divided ** 0.5
print("standard deviation: " + str(p2_standard_deviation))
```

```python
# cross verification

print(np.mean(p2_25000_samples1))

#  variance
print(np.std(p2_25000_samples1, ddof=1))
```
